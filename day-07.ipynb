```python
import aocd
import re
import numpy as np

%run helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=7)
data = puzzle.input_data
```

```python
grid_data = re.sub("S", "1", data)
grid_data = re.sub("\\.", "0", grid_data)
grid_data = re.sub("\\^", "2", grid_data)
grid = Grid(grid_data, conv_func=int)
grid[grid == 2] = -1

active_splitters = []
for i in range(len(grid)-1):
    beams = set(np.where(grid[i] > 0)[0])
    splitters = set(np.where(grid[i+1] == -1)[0])
    active_splitters += [(i+1, s) for s in beams & splitters]
    for beam in beams & splitters:
        grid[i+1][beam-1] += grid[i][beam]
        grid[i+1][beam+1] += grid[i][beam]
    for beam in beams - splitters:
        grid[i+1][beam] += grid[i][beam]
```

```python
puzzle.answer_a = len(active_splitters)
```
