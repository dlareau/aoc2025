```python
import aocd
import re
import numpy as np
%run helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=5)
data = puzzle.input_data
```

```python
ranges, items = data.split("\n\n")
items = list(map(int,items.split("\n")))
ranges = [list(map(int, x.split("-"))) for x in ranges.split("\n")]
```

```python
total = 0
for item in items:
    for r in ranges:
        if item in range(r[0], r[1]+1):
            total += 1
            break
puzzle.answer_a = total
```

```python
def add_range(ranges, r):
    start_range = -1
    end_range = -1
    duplicates = []
    for i in range(len(ranges)):
        if r[0] >= ranges[i][0] and r[1] <= ranges[i][1]:
            return
        if r[0] <= ranges[i][0] and r[1] >= ranges[i][1]:
            duplicates.append(i)
            continue
    
        if ranges[i][0] <= r[0] and r[0] <= ranges[i][1]:
            start_range = i
        if ranges[i][0] <= r[1] and r[1] <= ranges[i][1]:
            end_range = i

    if start_range != -1 and end_range == -1:
        ranges[start_range][1] = r[1]
    if start_range == -1 and end_range != -1:
        ranges[end_range][0] = r[0]
    if start_range != -1 and end_range != -1:
        ranges[start_range][1] = ranges[end_range][1]
        duplicates.append(end_range)

    for i in sorted(duplicates, reverse=True):
        del ranges[i]
        
    if start_range == -1 and end_range == -1:
        ranges.append(r[:])
        ranges.sort(key=lambda x:x[0])
    
```

```python
new_ranges = []
for r in ranges:
    add_range(new_ranges, r)
total = sum([r[1] - r[0] + 1 for r in new_ranges])
puzzle.answer_b = total
```
