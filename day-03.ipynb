```python
import aocd
import re
import numpy as np
%run helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=3)
data = puzzle.input_data
```

```python
banks = [list(map(int,list(x))) for x in data.split("\n")]
```

```python
# Original cursed 1-liner part A solution
puzzle.answer_a = sum([max(bank[:-1]) * 10 + max(bank[bank.index(max(bank[:-1]))+1:]) for bank in banks])
```

```python
def max_joltage(bank, digits):
    joltage = 0
    prev_index = -1
    for i in range(digits-1, -1, -1):
        eligible = bank[prev_index+1:len(bank)-i]
        j = max(eligible)
        prev_index += eligible.index(j) + 1
        joltage += j * (10 ** i)
    return joltage
```

```python
puzzle.answer_a = sum([max_joltage(bank, 2) for bank in banks])
```

```python
puzzle.answer_b = sum([max_joltage(bank, 12) for bank in banks])
```
