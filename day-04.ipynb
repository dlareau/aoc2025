```python
def conv(x):
    return 1 if x == "@" else 0

grid = Grid(data, conv_func=conv)
```

```python
total = 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == 1 and sum(grid.all_sides((i,j), vals=True)) < 4:
            total += 1
puzzle.answer_a = total
```

```python
grid2 = grid.copy()
total = 0
current_total = 1
while current_total != 0:
    current_total = 0
    for i in range(len(grid2)):
        for j in range(len(grid2[0])):
            if grid2[i][j] == 1 and sum(grid2.all_sides((i,j), vals=True)) < 4:
                current_total += 1
                grid2[i][j] = 0
    total += current_total
puzzle.answer_b = total
```
