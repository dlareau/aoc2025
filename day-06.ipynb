```python
import aocd
import re
import numpy as np
from math import prod

%run helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=6)
data = puzzle.input_data
```

```python
def do_math(operands, operator):
    return prod(operands) if operator == "*" else sum(operands)
```

```python
csv = re.sub(r" +", ",", data.strip())
grid = np.array([r.split(",") for r in csv.split("\n")]).transpose()

puzzle.answer_a = sum([do_math(map(int, x[:-1]), x[-1]) for x in grid])
```

```python
grid = [[x for x in row] for row in data.split("\n")]

operators = [x for x in grid.pop() if x != " "]

operands = ["".join(x).strip() for x in np.array(grid).transpose()]
operands = ",".join([x if x != "" else "-" for x in operands])
operands = [list(map(int,x.split(","))) for x in operands.split(",-,")]

operations = list(zip(operands, operators))
puzzle.answer_b = sum([do_math(x[0], x[1]) for x in operations])
```
