```python
import aocd
import re
import numpy as np
%run helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=2)
data = puzzle.input_data
```

```python
ranges = data.split(",")
ranges = [tuple(map(int, r.split("-"))) for r in ranges]
```

```python
total = 0
for r in ranges:
    for i in range(r[0], r[1]+1):
        divisor = (10 ** (len(str(i))/2) + 1)
        result = i / divisor
        if result == int(result):
            total += i
```

```python
puzzle.answer_a = total
```

```python
# Method 1, divisiblility test
length_map = {
    1: [],
    2: [11],
    3: [111],
    4: [1111,101],
    5: [11111],
    6: [111111,1001,10101],
    7: [1111111],
    8: [11111111,10001,1010101],
    9: [111111111,1001001],
    10: [1111111111,100001,101010101]
}

total2 = 0
for r in ranges:
    for i in range(r[0], r[1]+1):
        for d in length_map[len(str(i))]:
            result = i / d
            if result == int(result):
                total2 += i
                break
```

```python
# Method 2, String based test
lengths = {}
for i in range(11):
    lengths[i] = [x for x in range(2,i+1) if i%x == 0]
total2 = 0
for r in ranges:
    for i in range(r[0], r[1]+1):
        s = str(i)
        l = len(s)
        for d in lengths[l]:
            if s == d*s[:(l//d)]:
                total2 += i
                break
```

```python
puzzle.answer_b = total2
```
