```python
import aocd
import re
import numpy as np
%run aoc2025/helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=1)
data = puzzle.input_data
```

```python
dir = {"L": -1, "R": 1}
parsed = [dir[x[0]] * int(x[1:]) for x in data.split("\n")]
```

```python
pos = 50
zeros = 0
for d in parsed:
    pos = (pos + d) % 100
    if pos == 0:
        zeros += 1
puzzle.answer_a = zeros
```

```python
pos = 50
zeros = 0
for d in parsed:
    next = pos + d
    if next < 0:
        zeros += (next * -1) // 100
    if next > 0:
        zeros += next // 100
    if pos > 0 and next < 0:
        zeros += 1
    if pos < 0 and next > 0:
        zeros += 1
    if next == 0:
        zeros += 1
    pos = next % 100
puzzle.answer_b = zeros
```
