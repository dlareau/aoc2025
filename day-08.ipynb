```python
import aocd
import re
import numpy as np
from itertools import combinations
from math import prod
import networkx as nx

%run helper.ipynb
puzzle = aocd.models.Puzzle(year=2025, day=8)
data = puzzle.input_data
```

```python
points = [Point3D(tuple(map(int, x.split(",")))) for x in data.split("\n")]
```

```python
length_map = {}

for p1, p2 in combinations(points, 2):
    d = p1.s_dist(p2)
    length_map[d] = (p1, p2)

keys = sorted(length_map.keys())
```

```python
g = nx.Graph()
shortest = [length_map[k] for k in keys[:1000]]
for e in shortest:
    g.add_edge(e[0], e[1])
sizes = sorted([len(x) for x in list(nx.connected_components(g))])
puzzle.answer_a = prod(sizes[-3:])
```

```python
full_graph = nx.Graph()
for p in points:
    full_graph.add_node(p)

for k in keys:
    full_graph.add_edge(length_map[k][0], length_map[k][1], length=k)
    if nx.is_connected(full_graph):
        puzzle.answer_b = length_map[k][0].x * length_map[k][1].x
        break
```
